#### 24,画图，y=x*x 其中 x = np.arange(0, 100, 0.1) （提示这里用到  matplotlib.pyplot 库）

```python
import matplotlib.pyplot as plt
# 创建数据
x_24 = np.arange(0, 100, 0.1)
y_24 = x_24 * x_24

# 绘制图形
plt.figure(figsize=(8, 4))
plt.plot(x_24, y_24, label="y = x^2")
plt.title("Plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

#### 25.画图。画正弦函数和余弦函数， x = np.arange(0, 3 * np.pi, 0.1)(提示：这里用到 np.sin() np.cos() 函数和 matplotlib.pyplot 库)

```python
x_25 = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x_25)
y_cos = np.cos(x_25)

# 绘制图形
plt.figure(figsize=(8, 4))
plt.plot(x_25, y_sin, label="Sine")
plt.plot(x_25, y_cos, label="Cosine")
plt.title("Sine and Cosine Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
